```python
def list_to_time(time_list):
    if len(time_list) == 1:
        hours = time_list[0]
        minutes = 0
    else:
        hours, minutes = time_list

    hours = int(hours) % 24
    minutes = int(minutes) % 60
    return f"{hours:02}:{minutes:02}"


formatted_series = result.apply(list_to_time)
```

```python
pattern = r"(\d{1})\s*out of \d{1}\s*stars(\d{1,})\s*ratings"
text = df["stars"].apply(lambda x : re.findall(pattern, x))
```

```python
def extract_numbers(text):
  if text:
    return int(text[0][0]), int(text[0][1])
  else:
    return "Not rated yet"


text = text.apply(extract_numbers )
```

```python
df["Total_Rating"] = text.apply(lambda x: x[1])
df["stars"] = text.apply(lambda x: x[0])
```
